```python
def get_open_price(ticker):
    #stock = yf.Ticker(ticker)
    data = yf.download(ticker, period='1d')
    open_price = data['Open'][0]
    return open_price

def get_percent_change(ticker):
    stock = yf.Ticker(ticker)
    data = stock.history(start='2022-07-25', end='2023-07-25')
    percent_change = ((data['Open'][-1] - data['Open'][0]) / data['Open'][0]) * 100
    return percent_change

def get_percent_change_6_month(ticker):
    stock = yf.Ticker(ticker)
    data = stock.history(start='2023-01-25', end='2023-07-25')
    percent_change = ((data['Open'][-1] - data['Open'][0]) / data['Open'][0]) * 100
    return percent_change

def get_percent_change_3_month(ticker):
    stock = yf.Ticker(ticker)
    data = stock.history(start='2023-04-25', end='2023-07-25')
    percent_change = ((data['Open'][-1] - data['Open'][0]) / data['Open'][0]) * 100
    return percent_change

def get_percent_change_1_month(ticker):
    stock = yf.Ticker(ticker)
    data = stock.history(start='2023-06-25', end='2023-07-25')
    percent_change = ((data['Open'][-1] - data['Open'][0]) / data['Open'][0]) * 100
    return percent_change
```

```python
# Load the list of tickers from the CSV file (assuming 'tickers.csv' contains the tickers in the first column)
tickers_df = pd.read_csv('s&p500.csv')
tickers = tickers_df.iloc[:, 0]

# Create a new DataFrame to store the results
results_df = pd.DataFrame(columns=['Ticker', 'Open Price', 
                        'One Year Change',"One Year Return Percentile",
                        'Six Months Change',"Six Months Return Percentile",
                        'Three Months Change',"Three Months Return Percentile",
                        'One Month Change',"One Month Return Percentile",
                        'Number of Shares to Buy'])
results_df
```

```python

# Fetch data for each ticker and populate the results DataFrame
for ticker in tickers:
    open_price = get_open_price(ticker)
    percent_change = get_percent_change(ticker)
    percent_change_6_month = get_percent_change_6_month(ticker)
    percent_change_3_month = get_percent_change_3_month(ticker)
    percent_change_1_month = get_percent_change_1_month(ticker)
    results_df = results_df.append({
        'Ticker': ticker,
        'Open Price': open_price,
        'One Year Change': percent_change,
        'Six Months Change': percent_change_6_month,
        'Three Months Change': percent_change_3_month,
        'One Month Change': percent_change_1_month
        
    }, ignore_index=True)

hqm_df = results_df
```

```python
time_period = ["One Year","Six Months","Three Months","One Month"]

for row in hqm_df.index:
    for time in time_period:
        change_col = f"{time} Change"
        percentile_col = f"{time} Return Percentile"
        hqm_df.loc[row,percentile_col] = stats.percentileofscore(hqm_df[change_col], hqm_df.loc[row,change_col])
```

```python
hqm_df["HQM_Score"] = 0
```

```python
for row in hqm_df.index:
    momentum_percentiles = []
    for time in time_period:
        momentum_percentiles.append(hqm_df.loc[row, f"{time} Return Percentile"])
    hqm_df.loc[row,"HQM_Score"] = mean(momentum_percentiles)
```

```python
hqm_df.sort_values("HQM_Score",ascending=False,inplace=True)
hqm_df1 = hqm_df[:50]
hqm_df1.reset_index(inplace=True)
```

```python
# Define the Excel file name
excel_file_name = 'Recommended Trades.xlsx'

# Create a Pandas Excel writer using the openpyxl engine
writer = pd.ExcelWriter(excel_file_name, engine='openpyxl')

# Write the DataFrame to the Excel file
hqm_df1.to_excel(writer, index=False)
# Get the workbook and worksheet objects
workbook1  = writer.book
worksheet = writer.sheets['Sheet1']

# Create cell formatting
header_fill = PatternFill(start_color='0A0A23', end_color='0000FF', fill_type='solid')
header_font = Font(color='FFFFFF', bold=True)
currency_format = '[$$-409]#,##0.00'

# Format the header row (first row)
for col_num, column_letter in enumerate(worksheet.iter_cols(min_col=1, max_col=len(hqm_df1.columns)), 1):  # Adjust the index to start from 1
    for cell in column_letter:
        cell.fill = header_fill
        cell.font = header_font

# Format the second column as currency
for cell in worksheet['B'][1:]:  # Starting from the second row
    cell.number_format = currency_format

# Save the Excel file
writer.save()
```
