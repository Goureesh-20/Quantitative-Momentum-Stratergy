```python
# Load the list of tickers from the CSV file (assuming 'tickers.csv' contains the tickers in the first column)
tickers_df = pd.read_csv('s&p500.csv')
tickers = tickers_df.iloc[:, 0]

# Create a new DataFrame to store the results
results_df = pd.DataFrame(columns=['Ticker', 'Open Price', 'Percent Change to 1 Year Ago','Percent Change to 6 Months Ago','Percent Change to 3 Months Ago','Percent Change to 1 Month Ago','Number of Shares to Buy'])
```

```python

# Fetch data for each ticker and populate the results DataFrame
for ticker in tickers:
    open_price = get_open_price(ticker)
    percent_change = get_percent_change(ticker)
    percent_change_6_month = get_percent_change_6_month(ticker)
    percent_change_3_month = get_percent_change_3_month(ticker)
    percent_change_1_month = get_percent_change_1_month(ticker)
    results_df = results_df.append({
        'Ticker': ticker,
        'Open Price': open_price,
        'Percent Change to 1 Year Ago': percent_change,
        'Percent Change to 6 Months Ago': percent_change_6_month,
        'Percent Change to 3 Months Ago': percent_change_3_month,
        'Percent Change to 1 Month Ago': percent_change_1_month
        
    }, ignore_index=True)

```

```python
hqm_df = results_df
#hqm_df.insert(3, "One Year Return Percentile", 0)
#hqm_df.insert(5, "Six Months Return Percentile", 0)
#hqm_df.insert(7, "Three Months Return Percentile", 0)
#hqm_df.insert(9, "One Month Return Percentile", 0)
# Define a dictionary with the current column names as keys and the new column names as values

new_column_names = {
    'Percent Change to 1 Year Ago': 'One Year Change',
    'Percent Change to 6 Months Ago': 'Six Months Change',
    'Percent Change to 3 Months Ago': 'Three Months Change',
    'Percent Change to 1 Month Ago': 'One Month Change',
    'Number of Shares to Buy': 'New Number of Shares'
}

#new_column_names = {
#    'Siz Months Change': 'Six Months Change',
#}
# Rename the columns using the 'rename' method
hqm_df = hqm_df.rename(columns=new_column_names)
```

```python
hqm_df = hqm_df.iloc[:, :-9]
```

```python
time_period = ["One Year","Six Months","Three Months","One Month"]

for row in hqm_df.index:
    for time in time_period:
        change_col = f"{time} Change"
        percentile_col = f"{time} Return Percentile"
        hqm_df.loc[row,percentile_col] = stats.percentileofscore(hqm_df[change_col], hqm_df.loc[change_col])

hqm_df
```

```python
from statistics import mean
```

```python
hqm_df["HQM_Score"] = 0
```

```python
for row in hqm_df.index:
    momentum_percentiles = []
    for time in time_period:
        momentum_percentiles.append(hqm_df.loc[row, f"{time} Return Percentile"])
    hqm_df.loc[row,"HQM_Score"] = mean(momentum_percentiles)

hqm_df
```
